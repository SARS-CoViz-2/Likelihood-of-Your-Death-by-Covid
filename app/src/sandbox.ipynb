```python
import pandas as pd

with open('data/age_data.json','w+') as f:
  ages = pd.read_csv('data/age_data.csv', header=0, names=['value', 'cases', 'Case count', 'deaths', 'Death count', 'pop', 'dc_ratio'])
  ages = ages[['value', 'pop', 'cases', 'deaths', 'dc_ratio']]
  ages[['cases', 'deaths', 'dc_ratio']] = ages[['cases', 'deaths', 'dc_ratio']]*100
  ages = ages.append({'value':'-', 'pop':0, 'cases':0, 'deaths':0, 'dc_ratio':0},ignore_index=True)
  f.write(f'{{ "age_data": {ages.to_json(orient="records")} }}')
  f.close()

with open('data/sex_data.json','w+') as f:
  sexes = pd.read_csv('data/sex_data.csv', header=0, names=['value', 'cases', 'Case count', 'deaths', 'Death Range', 'Death count', 'pop', 'dc_ratio'])
  sexes = sexes[['value', 'pop', 'cases', 'deaths', 'dc_ratio']]
  sexes['pop'] = sexes['pop']
  sexes[['cases', 'deaths', 'dc_ratio']] = sexes[['cases', 'deaths', 'dc_ratio']]*10000
  sexes = sexes.append({'value':'-', 'pop':0, 'cases':0, 'deaths':0, 'dc_ratio':0},ignore_index=True)
  f.write(f'{{ "sex_data": {sexes.to_json(orient="records")} }}')
  f.close()


with open('data/ethnicity_data.json','w+') as f:
  races = pd.read_csv('data/ethnicity_data.csv', header=0, names=['value', 'cases', 'Case count', 'pop', 'deaths', 'Death count',  'dc_ratio'])
  races = races[['value', 'pop', 'cases', 'deaths', 'dc_ratio']]
  races[['cases', 'deaths', 'dc_ratio']] = races[['cases', 'deaths', 'dc_ratio']]*10000
  races = races.append({'value':'-', 'pop':0, 'cases':0, 'deaths':0, 'dc_ratio':0},ignore_index=True)
  f.write(f'{{ "race_data": {races.to_json(orient="records")} }}')
  f.close()
```

```python
import json
from string import ascii_lowercase as alpha

age_vals = pd.read_csv('data/age_data.csv', header=0, names=['value', 'cases', 'Case count', 'deaths', 'Death count', 'pop', 'dc_ratio'])['value'].values
sex_vals = pd.read_csv('data/sex_data.csv', header=0, names=['value', 'cases', 'Case count', 'deaths', 'Death Range', 'Death count', 'pop', 'dc_ratio'])['value'].values
race_vals = pd.read_csv('data/ethnicity_data.csv', header=0, names=['value', 'cases', 'Case count', 'deaths', 'Death count', 'pop', 'dc_ratio'])['value'].values
vals = [age_vals, sex_vals, race_vals]
labels = ['age','sex', 'race']

a = 0
fields = open('fields.json', 'w+')
fields.write(f'{{"options" :{{\n')
for i, (val, label) in enumerate(zip(vals,labels)):
  fields.write(f'   "{label}" : [\n')
  for v in val:
    fields.write(f'    {{"key" : "{alpha[a]}", "text": "{v}", "value": "{v}"}},\n')
    a += 1
  
  fields.write(f'    {{"key" : "{alpha[a]}", "text": "-", "value": "-"}}\n')
  a += 1
  
  if i == len(labels)-1:
    fields.write('  ]\n')
  else:
    fields.write('  ],\n')
fields.write('}}')
fields.close()
```
